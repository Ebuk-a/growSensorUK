```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
```

```python
grow_df = pd.read_csv('resources/GrowLocations.csv')
```

```python
grow_df
```

```python
filtered_grow_df = grow_df[(grow_df['Latitude'] >= 50.681) & (grow_df['Latitude'] <= 57.985) 
                        & (grow_df['Longitude'] >= -10.592) & (grow_df['Longitude'] <= 1.6848)]

filtered_grow_df
```

```python
geometry = [Point(xy) for xy in zip(filtered_grow_df['Longitude'], filtered_grow_df['Latitude'])]
gdf = GeoDataFrame(filtered_grow_df, geometry=geometry)  
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(10, 6)), marker='o', color='red', markersize=1)
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world[world.continent == 'Europe'].plot(figsize=(10, 6)), marker='o', color='red', markersize=15)
```

```python


df = pd.read_csv("Long_Lats.csv", delimiter=',', skiprows=0, low_memory=False)

geometry = [Point(xy) for xy in zip(df['Longitude'], df['Latitude'])]
gdf = GeoDataFrame(df, geometry=geometry)   

#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(10, 6)), marker='o', color='red', markersize=15)
```

```python
gdf = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"), where="continent='Europe'",
)
gdf.plot(ax=world.plot(figsize=(10, 6)), marker='o', color='red', markersize=15)
```
